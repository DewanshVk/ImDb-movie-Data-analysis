### Data Cleaning 

```python
# Droping some column from the DataFrame
movie.drop('img_link', axis=1, inplace=True)
movie.drop('cast_id', axis=1, inplace=True)
movie.drop('director_id', axis=1, inplace=True)
movie.drop('writter_id', axis=1, inplace=True)
```

```python
movie.columns
```

```python
# Sorting Values based on Duration
duration = []
for i in movie['duration']:
    if ('Nan' != i):
        duration.append(int(i))
    else:
        duration.append(100)
    
movie['duration'] = duration

movie.sort_values(by = 'duration', ascending = False).head()
```

```python
# Sorting the movies based on Year
movie.sort_values(by = 'year', ascending = False)
```

### Data visualization

 #### Q. How does the IMDb rating of movies in the top 250 list vary with their runtime?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movie, x='duration', y='imdb_rating', alpha=0.7)

# Add labels and title
plt.xlabel('Runtime ')
plt.ylabel('IMDB Rating')
plt.title('IMDB Rating vs. Movie Runtime')

# Show the plot
plt.show()
```

#### Q. What is the distribution of movie genres in the top 250 list?

```python
# Extract and count genres
genre_counts = movie['genre'].value_counts()

# Create a bar plot
plt.figure(figsize=(18, 6))
genre_counts.plot(kind='bar')

# Add labels and title
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.title('Distribution of Movie Genres in Top 250 List')

# Show the plot
plt.show()
```

#### Q. How does the number of votes (popularity) relate to the IMDb rating for the top 250 movies?

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movie, x='imbd_votes', y='imdb_rating', alpha=0.7)

# Add labels and title
plt.xlabel('Number of Votes')
plt.ylabel('IMDB Rating')
plt.title('IMDB Rating vs. Number of Votes')

# Show the plot
plt.show()
```

#### Q. How has the IMDb rating of movies evolved over the years?

```python
# Create a line plot
plt.figure(figsize=(10, 6))
movie.groupby('year')['imdb_rating'].mean().plot(marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Average IMDB Rating')
plt.title('Evolution of IMDb Rating Over the Years')

# Show the plot
plt.show()
```
